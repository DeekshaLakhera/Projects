## Popular50 Books

```python
#merge books and ratings datasets
```

```python
books_ratings=books.merge(ratings, on='ISBN')
books_ratings.head(1)
```

```python
# counting books reviews wrt to how many times it rated(reviewed by users)
```

```python
Total_Reviews=books_ratings.groupby('Book-Title').count()['Book-Rating'].reset_index().rename(columns={'Book-Rating':'Reviews-Counts'})
Total_Reviews.head()
```

```python
# Average rating in each book
```

```python
Ratings_mean=books_ratings.groupby('Book-Title')['Book-Rating'].mean().reset_index().rename(columns={'Book-Rating':'Avg-Ratings'})
Ratings_mean.head()
```

```python
# Top 50 books
```

```python
popular_50_books= Total_Reviews.merge(Ratings_mean,on='Book-Title')
popular_50_books.head()
```

```python
popular_50_books=popular_50_books[popular_50_books['Reviews-Counts']>=250].sort_values(by='Avg-Ratings',ascending=False).head(50)
popular_50_books.shape
```

```python
popular_50_books=popular_50_books.merge(books,on="Book-Title").drop_duplicates('Book-Title')[['Book-Title','Book-Author','Year-Of-Publication','Publisher','Image-URL-M','Reviews-Counts','Avg-Ratings']]
```

# Collaborative Filtering

```python
# Users who rates more than 200 books
```

```python
books_ratings=books.merge(ratings, on='ISBN')
books_ratings.head(1)
```

```python
books_ratings.head(1)
```

```python
# counting which user rates maximum number of books
```

```python
Reviewver=pd.DataFrame(books_ratings.groupby('User-ID').count()['Book-Rating']).rename(columns={'Book-Rating':'num_review'})
Reviewver=Reviewver[Reviewver['num_review']>200]
user_ids=Reviewver.index
user_ids
```

```python
# books rated by user_ids
```

```python
filtered_rating_data=books_ratings[books_ratings['User-ID'].isin(user_ids)]
```

```python
# count which books rated by user more than 50 times
```

```python
y=filtered_rating_data.groupby('Book-Title').count()['Book-Rating']>=50
famous_Books=y[y].index
```

```python
ratings_final=filtered_rating_data[filtered_rating_data['Book-Title'].isin(famous_Books)]
```

```python
ratings_final.duplicated().sum()
```

```python
ratings_final.head()
```

```python
pivot_t=ratings_final.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
```

```python
pivot_t.fillna(0,inplace=True)
```

```python
pivot_t
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score=cosine_similarity(pivot_t)
```

```python
similarity_score.shape
```

```python
def recommond(book_name):
    #index fetch of book_name
    index=np.where(pivot_t.index==book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_score[index])),key=lambda x:x[1],reverse=True)[1:6]
    for i in similar_items:
        print(pivot_t.index[i[0]])
```

```python
recommond('1984')
```

```python
recommond('You Belong To Me')
```
